```python
result = db.movies.aggregate([
    #資料篩選：match
    {
        "$match":{"type":"movie"}
    },
    #資料聚合與統計
    {
        "$group":{
            "_id":"$countries",
            "averageRuntime":{"$avg":"$runtime"}
        }
    }
])
```

```python
i = 0
for j in result:
    pprint.pprint(j)
    i+=1
    if i > 4:
        break
```

```python
result = db.movies.aggregate([
    {
        # unwind: 拆分。將陣列欄位的每一個值拆分為單獨的document
        "$unwind":"$genres"
    },
    {
        "$group":{
            "_id":"$genres",
            "count":{"$sum":1}
        }
    },
    {
        "$sort":{"count":-1}
    },
    {
        "$limit":1
    }
])

```

```python
db.movies.aggregate([
    {
        "$group": {
            "_id":"$year",
            "averageRating":{"$avg":"$imdb.rating"},
            "movies": {"$push":"$$ROOT"}
        }
    },
    {
        "$sort":{"averageRating":-1}
    },
    {
        "$limit":1
    },
    {
        "$unwind":"$movies"
    },
    {
        "$replaceRoot":{"newRoot":"$movies"}
    }
])
```
