```python
# Question 60 : find the number of restaurants that have been graded in each month of the year.
# The $unwind stage in aggregation pipeline, which is used to create a new document for each element in the grades array field
# The $project in aggregation pipeline is used to create new fields based on existing fields in the documents.
# The $toDate operator converts the date field in the grades subdocument to a date object, and then use the $month and $year operators extracts the month and year from the date object.
list(db.aggregate([
    {
        "$unwind":"$grades"
    },
    {
        "$project":
        {
            "month":{"$month":"$grades.date"},
            "year":{"$year":"$grades.date"}
        }
    },
    {
        "$group":
        {
            "_id":{"month":"$month","year":"$year"},
            "count":{"$sum":1}
        }
    },
    {
        "$sort":
        {
            "_id.year":1,
            "_id.month":1
        }
    }
    
]))
```

```python
# 85.  find the top 5 restaurants with the highest average score for each cuisine type, along with their average scores.
list(db.aggregate(
    [
        {
            "$unwind":"$grades"
        },
        {
            "$group":
            {
                "_id":{"cuisine":"$cuisine", "restaurant_id":"$restaurant_id"},
                "avgScore":{"$avg":"$grades.score"}
            }
        },
        {
            "$sort":{"_id.cuisine":1,"avgScore":-1}
        },
        {
            "$group":
            {
            "_id":"$_id.cuisine",
            "topRestaurants":{"$push":{"restaurant_id":"$_id.restaurant_id", "avgScore":"$avgScore"}}
            }
        },
        {
            "$project":{
                "_id":0,
                "cuisine":"$_id",
                "topRestaurants":{"$slice":["$topRestaurants",5]}
            }
        }


    ]
))
# 第一個sort 先按照cuisine排序（把同個cuisine的排在一起），在按照分數倒序排序
# aggregate 中的push：把原本分組前每一筆資料的特定欄位，加入陣列內來紀錄。
```

```python
# 86. find the top 5 restaurants in each borough with the highest number of "A" grades.
list(db.aggregate(
    [
        {
            "$unwind":"$grades"
        },
        {
            "$match":
            {
                "grades.grade":"A"
            }
        },
        {
            "$group":
            {
                "_id":{"borough":"$borough","restaurant_id":"$restaurant_id"},
                "count":{"$sum":1}
            }
        },
        {
            "$sort":
            {
                "_id.borough":1,
                "count":-1
            }
        },
        {
            "$group":
            {
                "_id":{"borough":"$_id.borough"},
                "topRestaurants":{"$push":{"restaurant_id":"$_id.restaurant_id", "count":"$count"}}
            }
        },
        {
            "$project":
            {
                "_id":0,
                "borough":"$_id",
                "topRestaurants":{"$slice":["$topRestaurants",5]}
            }
        }

    ]
))
```

```python
# 87. find the borough with the highest number of restaurants that have a grade of "A" and a score greater than or equal to 90.

list(db.aggregate(
    [
        {
            "$match":{
                    "grades.grade":"A",
                    "grades.score":{"$gte":90}
            }
        },
        {
            "$group":{
                "_id":{"borough":"$borough"},
                "count":{"$sum":1}
            }
        },
        {
            "$sort":{
                "count":-1
            }
        }
    ]
))
```
